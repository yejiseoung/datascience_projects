## Model

I used fastai library to create a model

```python
fruits = DataBlock(
    blocks = (ImageBlock, CategoryBlock),
    get_items = get_image_files,
    get_y = parent_label,
    splitter = RandomSplitter(seed=42),
    item_tfms = Resize(460),
    batch_tfms = aug_transforms(size=224, min_scale=0.75)
)
```

```python
fruits.summary(t_path)
```

```python
dls = fruits.dataloaders(t_path, bs=64)
```

```python
dls.show_batch()
```

```python
learn = cnn_learner(dls, resnet34, metrics=error_rate)
lr_min, lr_steep = learn.lr_find()
```

```python
print(f'Minimum/10: {lr_min:.2e}, steepest point: {lr_steep:.2e}')
```

```python
lr = 1e-2
```

```python
#learn.fit_one_cycle(5, base_lr=lr)
```

```python
learn.fine_tune(3, base_lr=lr)
```

```python
learn.loss_func
```

## Sketch CNN model

There are 4 types of layers in CNN model, and I created it step by step. 

### Convolutional layer

```python
def conv(ni, nf): return nn.Conv2d(ni, nf, kernel_size=3, stride=2, padding=1)
```

`ni` is equivalent to the depth of filter, and `nf` is equivalent to how many filters we will use. 

### Batch Normalization

It would be great if we normalize our data throughout the network, because it allows to improve the model with training speed and the model can learn each layer independently. 

```python
def bn(nf): return nn.BatchNorm2d(nf)
```

`nf` is equivalent to the filter output from the previous convolutional layer. 

### Activation functions_ReLU

Rectified Linear Unit (ReLU) turns any negative values to zero. Using ReLU as an activation function has several advantages like computational simplicity and representational sparsity, therefore in deep learning, ReLU became the default activation function (but probably not with RNNs).

```python
def ReLU(): return nn.ReLU(inplace=False)
```

### Flattening

The last thing we need to do is to take all these activations and outcoming matrix and flatten it into a single dimention of predictions. `Flatten()` does this performance. 

```python
Flatten??
```

### Making a Model

- Seven convolutional layers

```python
model = nn.Sequential(
    conv(3, 64),
    bn(64),
    ReLU(),
    conv(64, 128),
    bn(128),
    ReLU(),
    conv(128, 256),
    bn(256),
    ReLU(),
    conv(256, 512),
    bn(512),
    ReLU(),
    conv(512, 256),
    bn(256),
    ReLU(),
    conv(256, 128),
    bn(128),
    ReLU(),
    conv(128, 131),
    bn(131),
    Flatten()    
)
```

```python
ln = Learner(dls, model, loss_func=CrossEntropyLossFlat(), metrics=error_rate)
```

```python
ln.summary()
```

```python
ln.lr_find()
```

```python
ln.fine_tune(3, lr=1e-1)
```

### Simplify it

```python
def conv2(ni, nf): return ConvLayer(ni, nf, stride=2)
```

```python
net = nn.Sequential(
    conv2(3, 64),
    conv2(64, 128),
    conv2(128, 256),
    conv2(256, 512),
    conv2(512, 256),
    conv2(256, 128),
    conv2(128, 131),
    Flatten()
)
```

```python
ln = Learner(dls, net, loss_func=CrossEntropyLossFlat(), metrics=error_rate)
```

```python
ln.fine_tune(3, lr=1e-1)
```

### More Simplify it

```python
class ResBlock(Module):
    def __init__(self, nf):
        self.conv1 = ConvLayer(nf, nf)
        self.conv2 = ConvLayer(nf, nf)
        
    def forward(self, x): return x + self.conv2(self.conv1(x))
```

```python
net = nn.Sequential(
    conv2(3, 64),
    ResBlock(64),
    conv2(64, 128),
    ResBlock(128),
    conv2(256, 512),
    ResBlock(256),
    conv2(512, 256),
    ResBlock(512),
    conv2(256, 128),
    ResBlock(256),
    conv2(128, 131),
    Flatten()
)
```

```python
net
```

### Final simplified model

```python
def conv_and_res(ni, nf): return nn.Sequential(conv2(ni, nf), ResBlock(nf))
```

```python
net = nn.Sequential(
    conv_and_res(3, 64),
    conv_and_res(64, 128),
    conv_and_res(128, 256),
    conv_and_res(256, 512),
    conv_and_res(512, 256),
    conv_and_res(256, 128),
    conv2(128, 131),
    Flatten()
)
```

```python
ln = Learner(dls, net, loss_func=CrossEntropyLossFlat(), metrics=error_rate)
```

```python
ln.lr_find()
```

```python
ln.fine_tune(3, lr=1e-1)
```

```python
dl = ln.dls.test_dl(test_items, rm_type_tfms=1, bs=64)
preds, _ = ln.get_preds(dl=dl)
preds.shape
thresh=0.5
labelled_preds = [' '.join([ln.dls.vocab[i] for i,p in enumerate(pred) if p>thresh]) for pred in preds.numpy()]
print(labelled_preds[:5])
y_test = list(map(lambda x: x.parents[0].name, test_items))
results = pd.DataFrame({'target': y_test, 'pred': labelled_preds})
accuracy = results[results.target==results.pred].shape[0] / results.shape[0]
accuracy
```
